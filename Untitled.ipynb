```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Define the file path
file_path = 'Desktop/TEAM - Dataset/Workspace/train/dataset_boundary_analysis.csv'

# Load the dataset
try:
    df = pd.read_csv(file_path)
    
    # Display the first few rows to check column names
    print("Dataset Head:")
    print(df.head())
    
    # Replace 'your_column_name' with the actual name of the column you want to plot
    column_to_plot = 'your_column_name' 
    
    if column_to_plot in df.columns:
        # Create the distribution plot
        plt.figure(figsize=(10, 6))
        sns.histplot(df[column_to_plot], kde=True, color='skyblue')
        
        plt.title(f'Distribution of {column_to_plot}')
        plt.xlabel(column_to_plot)
        plt.ylabel('Frequency')
        plt.show()
    else:
        print(f"Error: Column '{column_to_plot}' not found in the dataset.")
        print(f"Available columns: {list(df.columns)}")

except FileNotFoundError:
    print(f"Error: The file at {file_path} was not found. Please check the file path.")
```
